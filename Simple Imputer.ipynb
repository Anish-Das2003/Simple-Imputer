```python
from sklearn.impute import SimpleImputer
```

```python
simple=SimpleImputer(strategy='mean')
```

```python
impute_cols=['Glucose','Diastolic_BP','Skin_Fold','Serum_Insulin','BMI']
```

```python
imputed_data=simple.fit_transform(df[impute_cols])
imputed_data
```

```python
df[impute_cols]=pd.DataFrame(imputed_data,columns=impute_cols)
```
