```python
df=pd.read_csv('https://raw.githubusercontent.com/611noorsaeed/100-days-Scikit-Learn-Tutorials-/refs/heads/main/9%20covid_toy.csv')
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.preprocessing import  OneHotEncoder, OrdinalEncoder
```

```python
from sklearn.impute import SimpleImputer
```

```python
transformer=ColumnTransformer(
    transformers=[
        ('tf1',SimpleImputer(strategy='mean'),['fever']),
        ('tf2',OrdinalEncoder(),['gender','cough']),
        ('tf3',OneHotEncoder(drop='first',sparse_output=False,dtype=int),['city'])
        ],
    remainder='passthrough'
)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['has_covid']=le.fit_transform(df['has_covid'])
```

```python
df.head()
```

```python
x=df.drop('has_covid',axis=1)
y=df['has_covid']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train_tf=transformer.fit_transform(x_train)
```

```python
x_test_tf=transformer.fit_transform(x_test)
```

```python
features_names=transformer.get_feature_names_out()
features_names
```

```python
x_train_tf=pd.DataFrame(x_train_tf,columns=features_names,index=x_train.index)
x_train_tf.head()
```
